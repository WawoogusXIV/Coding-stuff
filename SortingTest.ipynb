```python
class Solution:
    def sort012(self, arr):
        arr.sort()


class Solution:
    def binarysearch(self, arr, k):
        left = 0
        right = len(arr) - 1
        ans = -1
        while left <= right:
            mid = left + (right - left) // 2
            if arr[mid] == k:
                ans = mid
                right = mid - 1
            elif arr[mid] < k:
                left = mid + 1
            else:
                right = mid - 1
        return ans
    
#One of these took 5 seconds and the other took what felt like an hour, try to guess which
#I got stuck on the second one before realizing "oh fuck it starts in the middle" and adapting
#Interesting how you can just figure out what's wrong with your code by talking to yourself like a madman
```
